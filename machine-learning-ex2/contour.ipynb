```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# plot decision boundary
# manually generate grid and calculate score for every grid

step = 100
exam_grid = np.hstack((
    np.linspace(0, 100, step).reshape((-1,1)),
    np.linspace(0, 100, step).reshape((-1,1))
))

predicted_grid = np.zeros((step, step))

# calculates each y by looping through every combination of grid
for i in range(step):
    for j in range(step):
        predicted_grid[i][j] = i^2 * j^2

plt.figure(figsize=((20,15)))
plt.contour(exam_grid[:,0], exam_grid[:,1], predicted_grid, 50, cmap = 'RdGy')
```
