```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx,yy = np.meshgrid(np.array([int(x) for x in range(-100,101,1)]), np.array([int(y) for y in range(-101,100,1)]))
```

```python
x_min, x_max = -100, 100
y_min, y_max = -100, 100
step = 201

# create a grid mesh
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, step),
    np.linspace(y_min, y_max, step)
)

# ravel the grid to a list, then stack together
# grid = np.hstack((np.ones(xx.ravel().shape[0]).reshape(-1,1), xx.ravel().reshape(-1,1), yy.ravel().reshape(-1,1)))
grid = xx + yy/4

# f_grid = np.array(f_grid).reshape(xx.shape) # convert to np array, then reshape to x by x grid, for contour plot later
# print(predicted_grid)

plt.figure(figsize=((12,8)))
plot = plt.contour(xx, yy, grid, 10)
plt.clabel(plot, inline=1, fontsize=10)
cmap = 'BlGy'
plt.colorbar() # create colorbar for scale
```
